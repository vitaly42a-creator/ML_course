```python
# Практическая работа №2
# Студент: Манаков Виталий, группа 934-М2 инд.5
# Вариант: 25

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# Настройка для красивого вывода pandas, при выводе таблиц показывать все столбцы, не обрезая их.
pd.set_option('display.max_columns', None)

# Названия колонок для датасета Adult
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race', 'sex',
    'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'
]

# Загрузка данных с сайта UCI
# Ссылка на исходный датасет Adult из репозитория UCI
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
# try: пробует выполнить код; если возникнет ошибка, перейди в специальный блок except
try:
    # index_col=False ни один столбец не нужно делать индексом
    # skipinitialspace=True убираем пробелы сразу после запятых при чтении файла т.к в Adult после разделителей часто есть пробелы
    df = pd.read_csv(url, names=column_names, index_col=False, skipinitialspace=True)
    print("Данные успешно загружены с UCI.")
    # df.shape возвращает (количество строк, количество столбцов)
    print("Размер датасета:", df.shape)
# Если при загрузке данных возникла ошибка например нет интернета или сайт недоступен, выполнение перейдёт в блок, "as e" сохранить объект ошибки в переменную e
except Exception as e:
    print(f"Не удалось загрузить данные с UCI. Ошибка: {e}")
    print("Проверьте подключение к интернету или попробуйте позже.")
    # raise e повторно вызывает ошибку, чтобы программа не продолжала работу с несуществующими данными
    raise e

# Выбираем все столбцы строкового типа (object)
obj_cols = df.select_dtypes(include="object").columns

# Для каждого строкового столбца удаляем лишние пробелы в начале и в конце строки
# strip() убирает пробелы в начале строки, в конце строки
for c in obj_cols:
    df[c] = df[c].str.strip()

# В датасете Adult пропущенные значения обозначены символом "?"
# Заменяем их на специальное значение NaN (Not a Number) которое pandas распознаёт как пропуск
df = df.replace("?", np.nan)

print("\nПЕРВИЧНЫЙ АНАЛИЗ:")

print("\nПервые 5 строк данных:")
display(df.head())

print("\nИнформация о датасете:")
print(df.info())

# describe() считает основные статистики по числовым признакам
print("\nОсновные статистики числовых колонок:")
display(df.describe())
print("\nОсновные статистики категорий:")
display(df.describe(include='object'))

# df.isna() создаёт таблицу True/False: True там, где есть пропуск
# .sum() считает количество пропусков по каждому столбцу
# .sort_values(ascending=False) сортирует столбцы по убыванию числа пропусков
print("\nПропуски по столбцам:")
display(df.isna().sum().sort_values(ascending=False).head(10))
```

```python
# Задание 1: Сравнение стратегий импутации для категориального признака

# Т.к occupation категориальный признак, среднее и стандартное отклонение для него не рассчитываются
# Вместо этого изменение оценивается через частоты и доли категорий до и после обработки

# Выбираем категориальный признак, в котором есть пропуски
col_miss = "occupation"
# Считаем, сколько пропусков в выбранном столбце
missing_count = df[col_miss].isna().sum()
# Считаем долю пропусков в процентах len(df) общее количество строк в датасете
missing_percent = (missing_count / len(df)) * 100

print(f"Пропусков в '{col_miss}': {missing_count} ({missing_percent:.2f}%)")

# ИСХОДНОЕ РАСПРЕДЕЛЕНИЕ КАТЕГОРИЙ

# Cчитаем частоты категорий до групировки (включая NaN), берем топ-12
# value_counts() считает, сколько раз встречается каждая категория
# .fillna("NaN") заменяет пропуски на текст "NaN"
vc_before = df[col_miss].fillna("NaN").value_counts().head(12)

# СТРАТЕГИЯ 1: УДАЛЕНИЕ СТРОК С ПРОПУСКАМИ

# Удалить строки, где в этом столбце NaN
# dropna(subset=[col_miss]) удаляет строки,где в столбце occupation стоит NaN
# .copy() создаёт независимую копию очищенной таблицы, чтобы безопасно работать с ней дальше
df_drop = df.dropna(subset=[col_miss]).copy()

# Считаем частоты категорий после удаления строк с пропусками
vc_drop = df_drop[col_miss].value_counts()

# СТРАТЕГИЯ 2: ЗАПОЛНЕНИЕ ПРОПУСКОВ МОДОЙ

# Заполнить пропуски модой
# mode() находит наиболее часто встречающееся значение
# dropna=True исключает пропуски из поиска моды
# [0] берёт первое значение, если мод несколько
mode_value = df[col_miss].mode(dropna=True)[0]

# Создаём копию исходной таблицы для второго варианта обработки
df_mode = df.copy()

# fillna(mode_value) заменяет все NaN в выбранном столбце на моду
df_mode[col_miss] = df_mode[col_miss].fillna(mode_value)

# Считаем частоты категорий после заполнения пропусков модой
vc_mode = df_mode[col_miss].value_counts()

print(f"Мода для '{col_miss}': {mode_value}")

# Визуализация: Гистограммы по категориям в виде столбчатых диаграмм

# Создаем фигуру с 3 графиками, 1 строка, 3 столбца графиков
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# До обработки
# Для наглядности выводим NaN отдельным столбцом заменим NaN на подпись "NaN"
axes[0].bar(vc_before.index.astype(str), vc_before.values)
axes[0].set_title(f"До обработки\nВсего: {len(df)} строк")
axes[0].set_xlabel("Категория")
axes[0].set_ylabel("Количество")
# Поворачиваем подписи категорий на оси X на 70 градусов
axes[0].tick_params(axis='x', rotation=70)

# Удаление строк
axes[1].bar(vc_drop.index.astype(str), vc_drop.values)
axes[1].set_title(f"Удаление строк\nОсталось: {len(df_drop)} строк")
axes[1].set_xlabel("Категория")
axes[1].set_ylabel("Количество")
axes[1].tick_params(axis='x', rotation=70)

# Заполнение модой
axes[2].bar(vc_mode.index.astype(str), vc_mode.values)
axes[2].set_title(f"Заполнение модой\nМода: '{mode_value}'")
axes[2].set_xlabel("Категория")
axes[2].set_ylabel("Количество")
axes[2].tick_params(axis='x', rotation=70)

plt.tight_layout()
plt.show()

# Текстовый комментарий под графиками
# посчитаем доли категорий до и после обработки
# value_counts(normalize=True) возвращает не абсолютные частоты, а относительные доли (от 0 до 1)
orig_share = df[col_miss].value_counts(normalize=True, dropna=True)
# Доли после удаления строк
drop_share = df_drop[col_miss].value_counts(normalize=True)
# Доли после заполнения модой
mode_share = df_mode[col_miss].value_counts(normalize=True)
# Получаем долю моды до обработки
# .get(mode_value, 0) безопасно возвращает значение по ключу, а если ключа нет вернёт 0
mode_before = orig_share.get(mode_value, 0) * 100
# Получаем долю моды после заполнения модой
mode_after  = mode_share.get(mode_value, 0) * 100

# ТЕКСТОВЫЙ КОММЕНТАРИЙ ПОД ГРАФИКАМИ

# f""" ... """ — это многострочная f-строка,
# в которую можно подставлять значения переменных.
display(Markdown(
f"""
**Комментарий к графикам**

На первом графике показано исходное распределение категорий признака **{col_miss}**,
причём пропуски отображены отдельно как категория **NaN**.

На втором графике видно распределение после удаления строк с пропусками.
По смыслу структура распределения почти сохраняется, но общий объём данных уменьшается:
остаётся **{len(df_drop)}** строк вместо **{len(df)}**.

На третьем графике показано распределение после заполнения пропусков модой.
В этом случае строки не теряются, но увеличивается частота самой популярной категории **{mode_value}**.

- Доля моды до обработки: **{mode_before:.2f}%**
- Доля моды после заполнения модой: **{mode_after:.2f}%**
"""
))

# Итоговый вывод
print("\nВЫВОД ПО ЗАДАНИЮ 1:")

# Показываем долю пропусков в исходном признаке
print(f"1. Доля пропусков в '{col_miss}': {missing_percent:.2f}%")

# При удалении строк теряем часть наблюдений
# Это может быть нежелательно, если данных не очень много
print(f"2. Удаление строк: потеряно {missing_count} строк ({missing_percent:.2f}% данных)")

# При заполнении модой строки сохраняются, но искусственно увеличивается частота самой популярной категории
print(f"3. Заполнение модой: добавлено {missing_count} записей категории '{mode_value}'")

# Итоговая интерпретация
print("\nДля категориальных признаков с небольшой долей пропусков оба метода часто дают похожую картину.")
print("fillna(мода) лучше, если важно не терять строки, например при обучении модели.")
print("dropna может быть уместен, если строк с пропусками немного или такие строки считаются нерелевантными по смыслу.")
print("Заполнение модой сохраняет объём данных, но сильнее искажает распределение, увеличивая долю самой частой категории.")
```

Задание 2: Детекция и визуализация выбросов
Цель: Научиться находить и аргументированно удалять аномалии. Инструкция:

Выберите числовой признак, наиболее важный для вашей задачи (по корреляции с таргетом).
Найдите выбросы методом IQR и методом Z-Score.
Визуализуйте их на Boxplot и Scatterplot (подсветите выбросы красным цветом).
Посчитайте процент выбросов от общего числа данных.
Вывод: Стоит ли удалять эти строки или они содержат ценную информацию (например, дорогие дома или редкие случаи болезни)?
Подсказка: Если выбросов > 5%, удаление может привести к потере информации. Рассмотрите возможность "обрезки" (capping) значений вместо удаления.

```python
# Задание 2: Детекция и визуализация выбросов

# В датасете Adult целевая переменная income является категориальной:
# например, значения могут быть "<=50K" ">50K"
# Чтобы считать корреляцию с числовыми признаками, удобно преобразовать target в бинарный вид: >50K  -> 1, <=50K -> 0
# .astype(str) приводим значения к строковому типу
# .str.contains(">50K") получаем True там, где строка содержит ">50K"
# .astype(int) True превращается в 1, False — в 0
df["income_bin"] = df["income"].astype(str).str.contains(">50K").astype(int)

# Выбираем все числовые столбцы из DataFrame
# select_dtypes(include=[np.number]) оставляет только числовые типы данных
# .columns.tolist() превращает названия столбцов в обычный Python-список
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Убираем из этого списка сам таргет income_bin, потому что он нужен только как переменная с которой сравниваем признаки, а анализировать выбросы в нём не нужно
num_cols = [c for c in num_cols if c != "income_bin"]

# corrwith(df["income_bin"]) считает корреляцию каждого числового признака со столбцом income_bin
# sort_values(key=lambda s: s.abs(), ascending=False) сортирует результаты по модулю корреляции
# Считаем корреляцию каждого числового признака с таргетом (income_bin)
corrs = df[num_cols].corrwith(df["income_bin"]).sort_values(key=lambda s: s.abs(), ascending=False)

print("\nВыбор числового признака по корреляции с таргетом:")
print("Топ-5 корреляций (по модулю):")
display(corrs.head(5))

# Берём первый признак из отсортированного списка это признак с самой сильной по модулю корреляцией с таргетом.
feature = corrs.index[0]
print(f"Выбран признак для анализа выбросов: {feature} (corr = {corrs.loc[feature]:.3f})")

# Берём выбранный столбец и приводим его к числовому типу
# errors="coerce" означает: если встретится что-то, что не удаётся преобразовать в число, оно будет превращено в NaN
x = pd.to_numeric(df[feature], errors="coerce")

# Удаляем пропуски из x и оставляем только валидные числовые значения
# Это нужно, чтобы статистики (квартили, среднее, std) считались только по реальным данным, а не по пропускам
x_valid = x.dropna()

# Метод 1: IQR Метод межквартильного размаха
# Метод IQR считает выбросами значения, которые слишком далеко находятся от "центральной" части распределения

# Сначала считаем:
# Q1 — первый квартиль (25% значений ниже него)
# Q3 — третий квартиль (75% значений ниже него)
q1 = x_valid.quantile(0.25)
q3 = x_valid.quantile(0.75)

# Межквартильный размах IQR = Q3 - Q1
# Он показывает ширину центральных 50% данных
iqr = q3 - q1

# Нижняя и верхняя границы выбросов
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

# Создаём булеву маску True если значение меньше lower или больше upper, то есть является выбросом по IQR
mask_iqr = (x < lower) | (x > upper)

# В маске могут быть NaN там, где x само было NaN
# fillna(False) заменяет такие случаи на False пропущенные значения не считаем выбросами
mask_iqr = mask_iqr.fillna(False)

# Суммируем True в маске True считается как 1, False как 0
# Получаем общее количество выбросов по IQR
iqr_outliers = int(mask_iqr.sum())

# Считаем долю выбросов среди валидных числовых значений
iqr_percent = iqr_outliers / len(x_valid) * 100

# Метод 2: Z-score Количество стандартных отклонений от среднего
# Метод Z-score оценивает, насколько далеко каждое значение находится от среднего в единицах стандартного отклонения

# Сначала считаем среднее и стандартное отклонение
mean_x = x_valid.mean()
std_x = x_valid.std()

# Рассчитываем z-score для каждого значения
z = (x - mean_x) / std_x
# Обычный порог если |z| > 3, то значение считается выбросом
mask_z = z.abs() > 3

# Там, где исходное значение было NaN, z тоже будет NaN
# Такие случаи не считаем выбросами
mask_z = mask_z.fillna(False)

# Считаем количество и процент выбросов по Z-score
z_outliers = int(mask_z.sum())
z_percent = z_outliers / len(x_valid) * 100

print("\nРезультаты детекции выбросов:")
print(f"IQR-границы: [{lower:.3f}; {upper:.3f}]")
print(f"Выбросов по IQR: {iqr_outliers} ({iqr_percent:.2f}%)")
print(f"Выбросов по Z-score (|z|>3): {z_outliers} ({z_percent:.2f}%)")

# Визуализация: Boxplot + Scatterplot (выбросы красным)

# Для scatterplot удобно использовать номер строки как ось X, это позволяет увидеть, где по порядку расположены выбросы
idx = np.arange(len(df))

fig, axes = plt.subplots(2, 2, figsize=(16, 8))

# Строим boxplot только по валидным значениям x_valid
sns.boxplot(x=x_valid, ax=axes[0, 0])
# Поверх boxplot наносим красные точки выбросы по IQR
# x[mask_iqr] значения, которые признаны выбросами
# np.zeros(iqr_outliers) координата по оси Y для этих точек
axes[0, 0].scatter(x[mask_iqr], np.zeros(iqr_outliers), color="red", s=20, label="IQR outliers")
axes[0, 0].set_title(f"Boxplot: {feature} (IQR)")
axes[0, 0].legend()

# Scatterplot + IQR выбросы красным

# Сначала рисуем все точки, которые НЕ являются выбросами по IQR
axes[0, 1].scatter(idx[~mask_iqr], x[~mask_iqr], alpha=0.4)
# Затем поверх рисуем выбросы красным цветом
axes[0, 1].scatter(idx[mask_iqr], x[mask_iqr], color="red", alpha=0.8, label="IQR outliers")
axes[0, 1].set_title(f"Scatter: {feature} (IQR)")
axes[0, 1].set_xlabel("Номер строки (индекс)")
axes[0, 1].set_ylabel(feature)
axes[0, 1].legend()

# Boxplot + Z-score выбросы красным

sns.boxplot(x=x_valid, ax=axes[1, 0])
# Накладываем поверх точки, признанные выбросами по Z-score
axes[1, 0].scatter(x[mask_z], np.zeros(z_outliers), color="red", s=20, label="Z outliers")
axes[1, 0].set_title(f"Boxplot: {feature} (Z-score)")
axes[1, 0].legend()

# Scatterplot + Z-score выбросы красным

axes[1, 1].scatter(idx[~mask_z], x[~mask_z], alpha=0.4)
axes[1, 1].scatter(idx[mask_z], x[mask_z], color="red", alpha=0.8, label="Z outliers")
axes[1, 1].set_title(f"Scatter: {feature} (Z-score)")
axes[1, 1].set_xlabel("Номер строки (индекс)")
axes[1, 1].set_ylabel(feature)
axes[1, 1].legend()

plt.tight_layout()
plt.show()

# Проценты выбросов от общего числа строк

# По условию задания нужно оценить процент выбросов от общего числа наблюдений
total_n = len(df)
# x.notna().sum() количество строк, где значение признака не пропущено
# Необходимо для вывода по скольким наблюдениям реально шёл расчёт
valid_n = x.notna().sum()

# Процент выбросов от общего числа строк
iqr_percent_total = iqr_outliers / total_n * 100
z_percent_total = z_outliers / total_n * 100

display(Markdown(
f"""
**Комментарий к графикам.**

Выбранный признак: **{feature}**
(корреляция с таргетом income_bin = **{corrs.loc[feature]:.3f}**).

**Boxplot (IQR):** красные точки - значения ниже **{lower:.3f}** или выше **{upper:.3f}**, то есть выбросы по методу IQR.
**Scatter (IQR):** красные точки показывают, что выбросы расположены на краях распределения и представляют экстремально большие или малые значения.
**Boxplot / Scatter (Z-score):** красные точки наблюдения с **|z| > 3**, то есть значения, сильно удалённые от среднего.

**Процент выбросов:**

IQR: **{iqr_outliers}** выбросов - **{iqr_percent_total:.2f}%** от всех строк (валидных значений: **{valid_n}**)

Z-score: **{z_outliers}** выбросов - **{z_percent_total:.2f}%** от всех строк
"""
))

print("\nВЫВОД ПО ЗАДАНИЮ 2:")

if iqr_percent_total > 5 or z_percent_total > 5:
    print("Выбросов довольно много (>5%). Удаление строк может привести к потере информации.")
    print("Лучше рассмотреть capping (обрезку значений) вместо удаления.")
else:
    print("Доля выбросов небольшая (<=5%). Удаление возможно, но стоит проверить смысл выбросов.")
    print("Если это реальные редкие случаи (например, очень большие значения), их удалять не всегда правильно.")
```

Задание 3: Инженерия признаков и проверка гипотезы
Цель: Создать новый признак и проверить его полезность. Инструкция:

Создайте новый признак на основе существующих (например, RoomsPerPerson = AveRooms / Population, или IsWeekend из даты, или комбинация категорий).
Постройте корреляционную матрицу с включением нового признака.
Постройте график зависимости нового признака от целевой переменной.
Вывод: Увеличил ли новый признак корреляцию с целевой переменной по сравнению с исходными? Рекомендуете ли вы его для модели?

```python
# Задание 3: Инженерия признаков и проверка гипотезы

# Проверяем, не был ли столбец income_bin уже создан раньше,чтобы не создавать его повторно
# >50K -> 1, <=50K -> 0
if "income_bin" not in df.columns:
    df["income_bin"] = (df["income"] == ">50K").astype(int)

# Специфическое требование: группируем редкие категории в признаке native-country
# Задаём порог редкости
col_country = "native-country"
min_count = 100

# Считаем частоты стран без пропусков
country_counts = df[col_country].dropna().value_counts()

# Определяем редкие страны оставляем только те, у которых число наблюдений меньше min_count, .index возвращает названия этих стран
rare_countries = country_counts[country_counts < min_count].index

# Создаём новый столбец native_country_grp, в котором будет сгруппированная версия исходного признака
# fillna("Missing") заменяет пропуски на отдельную категорию "Missing"
df["native_country_grp"] = df[col_country].fillna("Missing")
# Объединяем все редкие страны в одну общую категорию "Other"
# isin(rare_countries) проверяет, входит ли значение в список редких стран
df.loc[df["native_country_grp"].isin(rare_countries), "native_country_grp"] = "Other"

print("Топ-10 категорий после группировки native-country:")
display(df["native_country_grp"].value_counts().head(10))

# Создаём новый признак
# признак capital_net может быть лучше связан с уровнем дохода,чем признаки capital-gain и capital-loss по отдельности
df["capital_net"] = df["capital-gain"].fillna(0) - df["capital-loss"].fillna(0)
new_feature = "capital_net"

# Строим корреляционную матрицу для признаков которые нужны для проверки гипотезы
corr_matrix = df[["income_bin", "capital-gain", "capital-loss", "capital_net"]].corr(numeric_only=True)

plt.figure(figsize=(8, 6))
# Строим тепловую карту корреляций
# annot=True показываем числовые значения внутри ячеек
# fmt=".2f" округляем до 2 знаков после запятой
# cmap="coolwarm" цветовая палитра
# square=True делаем ячейки квадратными
# linewidths=0.5 тонкие линии между ячейками
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True, linewidths=0.5)
plt.title("Корреляционная матрица")
plt.tight_layout()
plt.show()

# График зависимости нового признака от целевой переменной

# Считаем статистики нового признака по классам income_bin
group_stats = df.groupby("income_bin")[new_feature].agg(["count", "mean", "median"])

print("\nСтатистика нового признака по классам income_bin:")
display(group_stats)

# Boxplot распределения нового признака по классам таргета

# Отдельно берём значения нового признака для каждой группы дохода
# income_bin = 0 - доход <= 50K
# income_bin = 1 - доход > 50K
vals0 = df.loc[df["income_bin"] == 0, new_feature].dropna()
vals1 = df.loc[df["income_bin"] == 1, new_feature].dropna()

plt.figure(figsize=(7, 4))
plt.boxplot([vals0, vals1], tick_labels=["<=50K (0)", ">50K (1)"], showfliers=True)
plt.title(f"Распределение {new_feature} по классам income_bin")
plt.ylabel(new_feature)
plt.tight_layout()
plt.show()

# Проверяем, улучшил ли новый признак корреляцию с таргетом

# Для проверки гипотезы сравним корреляцию income_bin: с capital-gain, с capital-loss, с capital_net
# Берём из матрицы корреляций только связи признаков с income_bin
base_feats = ["capital-gain", "capital-loss", new_feature]

# Считаем корреляционную матрицу только для этих признаков и income_bin
# Берём из матрицы корреляций только связи признаков с income_bin и удаляем сам income_bin из результата
base_corrs = df[base_feats + ["income_bin"]].corr(numeric_only=True)["income_bin"].drop("income_bin")

print("\nСравнение корреляций с income_bin:")
display(base_corrs)

# Среди исходных признаков (без нового) выбираем лучший по модулю корреляции
best_base = base_corrs.drop(new_feature).abs().max()

# Берём модуль корреляции нового признака
new_corr_abs = abs(base_corrs[new_feature])

print("\nВЫВОД ПО ЗАДАНИЮ 3:")

if new_corr_abs > best_base:
    print(
        f"Новый признак '{new_feature}' усилил связь с таргетом: |corr| = {new_corr_abs:.4f}, "
        f"что лучше, чем у исходных признаков ({best_base:.4f})."
    )
    print("Рекомендация: признак можно использовать в модели.")
else:
    print(
        f"Новый признак '{new_feature}' не улучшил корреляцию: |corr| = {new_corr_abs:.4f}, "
        f"а лучший из исходных признаков имеет |corr| = {best_base:.4f}."
    )
    print("Рекомендация: по линейной корреляции признак не даёт преимущества, но может быть полезен в нелинейных моделях.")
```
